```python
import pandas as pd
from sklearn.model_selection import train_test_split
df = pd.read_csv("Recent_UpdatedNB15.csv")
```

```python
df.shape
```

```python
local_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
local_df.to_csv("local_data.csv", index=False)
test_df.to_csv("test_data.csv", index=False)
```

```python
# Analyze binary class distribution in the datasets
print("=== BINARY CLASS DISTRIBUTION ANALYSIS ===")
print(f"Original dataset shape: {df.shape}")
print(f"Local training data shape: {local_df.shape}")
print(f"Test data shape: {test_df.shape}")

# Check binary class distribution
print("\n--- Original Dataset ---")
if 'label' in df.columns:
    orig_binary_dist = df['label'].value_counts().sort_index()
    print(f"Binary distribution: {dict(orig_binary_dist)}")
    print(f"Imbalance ratio: {orig_binary_dist.max() / orig_binary_dist.min():.2f}:1")

print("\n--- Local Training Data ---")
if 'label' in local_df.columns:
    local_binary_dist = local_df['label'].value_counts().sort_index()
    print(f"Binary distribution: {dict(local_binary_dist)}")
    print(f"Imbalance ratio: {local_binary_dist.max() / local_binary_dist.min():.2f}:1")

print("\n--- Test Data ---")
if 'label' in test_df.columns:
    test_binary_dist = test_df['label'].value_counts().sort_index()
    print(f"Binary distribution: {dict(test_binary_dist)}")
    print(f"Imbalance ratio: {test_binary_dist.max() / test_binary_dist.min():.2f}:1")

# Check attack type distribution
print("\n=== ATTACK TYPE DISTRIBUTION ===")
if 'attack_cat' in df.columns:
    print("Attack categories distribution:")
    attack_dist = df['attack_cat'].value_counts().sort_index()
    for attack_type, count in attack_dist.items():
        percentage = (count / len(df)) * 100
        print(f"  {attack_type}: {count} ({percentage:.2f}%)")
```
